# Lets try Ridge Regression on callifornia houseing price

```python
# import dataset
from sklearn.datasets import fetch_california_housing

# libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
```

```python
california_housing = fetch_california_housing()
california_housing
```

```python

# the dataset has no missing values
california_housing_df = pd.DataFrame(california_housing["data"], columns=california_housing["feature_names"])
california_housing_df
```

```python
california_housing_df["target"] = california_housing["target"]
california_housing_df.info()
```

```python
# features and target
X = california_housing_df.drop("target", axis=1)
y = california_housing_df["target"]
```

```python
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# initiate model
ridge = Ridge()
ridge.fit(X_train, y_train)
```

```python
# score
ridge.score(X_test, y_test)
```
