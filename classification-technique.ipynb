# Explore sklearn classification techniques

```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
heart_diseas = pd.read_csv("/kaggle/input/heart-disease-example/heart-disease.csv")
heart_diseas.head()
```

```python
heart_diseas.info()
```

```python
# Split into features and target
X = heart_diseas.drop("target", axis=1)
y = heart_diseas["target"]
```

```python
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

## 1.1 We'll try with `LinearSVC` model from `svm` module

```python
# import model
from sklearn.svm import LinearSVC
```

```python
linear_svc = LinearSVC(max_iter=10000)
```

```python
# fit the model and predict
linear_svc.fit(X_train, y_train)
linear_svc.score(X_test, y_test)
```

### So the above model `LinearSVC` from `svc` module seems like didn't satisfy need, so lets try different model

## 1.2 We'll try different `naive_bayes` module's algorithm (model)
* GaussianNB
* MultinomialNB
* ComplementNB
* BernoulliNB
* CategoricalNB

```python
# Here the step
# 1. import all the models
# 2. make a dictionary of models
# 3. loop through model dictionary
# 4. fit the model and predict
# 5. print the prediction
```

```python
# 1. Import all the models
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
```

```python
# 2. make a dictionary of all models
models = {
    "gaussian" : GaussianNB,
    "multinomia" : MultinomialNB,
    "complement" : ComplementNB,
    "bernoulli" : BernoulliNB,
}
```

```python
# 3. loop through models
model_instances = {}
for model_name, model_class in models.items():
    model_instances[model_name] = model_class()
model_instances
```

```python
# 4. fit the model and predict
for model_name , model_instance in model_instances.items():
    model_instance.fit(X_train, y_train)
    y_pred = model_instance.predict(X_test)  # Use predict to get predictions
    accuracy = (y_pred == y_test).mean()  # Calculate accuracy
    print(f"Accuracy for {model_name}:", accuracy)
```

## So from the above model `bernoulli` predict well

## 1.3 Lets try `RandomForestClassifier` model from `ensemble` module

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# fit the model and predict
random_forest_classifier = RandomForestClassifier()
random_forest_classifier.fit(X_train, y_train)
random_forest_classifier.score(X_test, y_test)
```

### Again `RandomForestClassifier` did good job than `LinearSVC`
